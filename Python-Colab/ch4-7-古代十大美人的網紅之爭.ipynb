```python
def swap(a, x, y):
    t = a[x]
    a[x] = a[y]
    a[y] = t

def comp(vt, r):
    for j in range(r):                    
        if vt[j] > vt[j+1]:               # 若左數 > 右數，則交換左右兩數
            swap(vt, j, j+1)              # 交換兩數(大的右移)

vote = [78,62,86,70,12,16,38,15,22,13]
c = len(vote)                 
for i in range(1,c):                      # c 個待排數字，只需排c-1回合
    comp(vote, c-i)                       # 每回合要完成「重複相鄰兩數比較與交換」的動作
    print('第', i, '回合結果:',vote)        # 印出每回合的結果

print('排序結束')
```

```python
def swap(a, x, y):
    t = a[x]
    a[x] = a[y]
    a[y] = t

def comp(bt, vt, r):
    for j in range(r):                    
        if vt[j] > vt[j+1]:               # 若左數 > 右數，則交換左右兩數
            swap(bt, j, j+1)              # 交換兩數(大的右移)
            swap(vt, j, j+1)

beauty = ['西施','貂蟬','王昭君','楊貴妃','馮小憐','蘇妲己','趙飛燕','鄭旦','褒姒','甄宓']
vote = [78,62,86,70,12,16,38,15,22,13]
c = len(vote)                 
for i in range(1, c):                     # c 個待排數字，只需排 c-1 回合
    comp(beauty, vote, c-i)               # 每回合要完成「重複相鄰兩數比較與交換」的動作  
    print('第', i, '回合結果:', beauty)     # 印出每回合的結果

print('排序結束')
```
