```python
score = [0,0,0,0,0]
for i in range(len(score)):
  score[i] = int(input())
  if score[i] < 70:
    print('可')
  elif score[i] < 90:
    print('良')
  else:
    print('優')
```

```python
score = [0,0,0,0,0]
found = 0
for i in range(len(score)):
  score[i] = int(input())
  if score[i] < 70:
    print('可')
  elif score[i] < 90:
    print('良')
  else:
    print('優')
    found = found + 1

print('共', found, '筆達到「優」等')
```
