```python
# 存在語意錯誤的程式碼
score = [83,92,76,96]
for i in range(4):
   sum = 0
   sum = sum + score[i]
print(sum)
```

```python
# 修正後的程式碼
score = [83,92,76,96]
sum = 0
for i in range(4):
   sum = sum + score[i]
print(sum)
```

```python
# 使用者輸入五個數字，最後程式輸出這五個數字的平均值
score = [0,0,0,0,0]
for i in range(len(score)):
  score[i] = int(input())

sum = 0
for i in range(len(score)):
   sum = sum + score[i]

print('平均:', sum / len(score))
```
